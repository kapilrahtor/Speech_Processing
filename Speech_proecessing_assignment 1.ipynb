Generate an audio file from text mentioning words such as, Sesquipedalian,
Perspicacity, Ostentation, Muliebrity,
Hippopotomonstrosesquipedaliophobia, Floccinaucinihilipilification,
Scripturient.

```python
from gtts import gTTS
import os
```

```python
test = ['Sesquipedalian', 'Perspicacity', 'Ostentation', 'Muliebrity', 'Hippopotomonstrosesquipedaliophobia', 'Floccinaucinihilipilification', 'Scripturient']
```

```python
test
```

```python
language = 'en'
```

```python
test[0]
```

```python
for i in test:
    print(i)
```

```python
for i in test:
    gtts_object = gTTS(text = i, lang = language,slow = False)
    gtts_object.save(str(i)+'.mp3')
```

```python
import librosa
import matplotlib.pyplot as plt
import librosa.display
X = []
SR = []
for i in test:
    audio_path = str(i)+'.mp3'
    x,sr = librosa.load(audio_path)
    plt.figure(figsize=(14,7))
    plt.title(str(i))
    librosa.display.waveshow(x,sr = sr)
    X  = librosa.stft(x)
    Xdb = librosa.amplitude_to_db(abs(X))
    plt.figure(figsize=(14,7))
    plt.title('Spectrum of'+' '+str(i))
    librosa.display.specshow(Xdb,sr = sr, x_axis = 'time', y_axis='hz')
    plt.colorbar()
```
