```python
import librosa
import librosa.display
from sklearn.preprocessing import minmax_scale
import IPython.display as ipd
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Happy

```python
happy= ['03-01-03-01-01-01-01.wav','03-01-03-02-01-01-01.wav','03-01-03-02-02-02-01.wav']
j = 1
for i in happy:
  x_w, sr_w = librosa.load(i)
  plt.figure(figsize = (20,5))
  plt.subplot(3,1,j)
  plt.title('Waveplot of' + i)
  librosa.display.waveshow(x_w, sr = sr_w)
  plt.show()
  j+=1
```

```python

happy= ['03-01-03-01-01-01-01.wav','03-01-03-02-01-01-01.wav','03-01-03-02-02-02-01.wav']
j = 1
for i in happy:
  x, sr = librosa.load(i)
  Xf = librosa.stft((x))

  Xdb = librosa.amplitude_to_db(abs(Xf))

  #plt.figure(figsize = (20,5))
  plt.subplot(3,1,j)
  plt.title('Log Spectrum of' + i)
  librosa.display.specshow(Xdb, sr = sr,x_axis = 'time' , y_axis = 'log')
  plt.show()
  j+=1
```

Zero Crossing

```python
from librosa.core.audio import zero_crossings
happy= ['03-01-03-01-01-01-01.wav','03-01-03-02-01-01-01.wav','03-01-03-02-02-02-01.wav']
j = 1
for i in happy:
  x, sr = librosa.load(i)

  zero_crossings = librosa.zero_crossings(x)
  print(' sum of the zero crossing of '+i,zero_crossings.sum())
  plt.figure(figsize=(20,5))
  plt.title('Zero Crossing Rate of' + i)
  zcrs = librosa.feature.zero_crossing_rate(x)
  plt.plot(zcrs[0])
  plt.show()
```

```python
from sklearn.preprocessing import minmax_scale
def normalize(x, axis = 0):
  return minmax_scale(x,axis=axis)
```

```python
happy = ['03-01-03-01-01-01-01.wav','03-01-03-02-01-01-01.wav','03-01-03-02-02-02-01.wav']
j = 1
for i in happy:
  x, sr = librosa.load(i)
  plt.figure(figsize=(20,5))
  plt.title('Spectral centroid of'+i)
  spectral_centroids = librosa.feature.spectral_centroid(y=x,sr= sr)[0]
  frames = range(len(spectral_centroids))
  t = librosa.frames_to_time(frames)
  librosa.display.waveshow(x,sr = sr, alpha = 0.4)
  plt.plot(t,normalize(spectral_centroids), color = 'r')
  plt.show()
```
